# Import the Packages

```python
from spi import *
import matplotlib.pyplot as plt
```

# Open precipitation file

```python
df = pd.read_csv('alcomat_chirps.csv',sep=',')
```

```python
# for calculate SPI we net to have a column with precipitation
# and a column with dates
# we can use the date column as index
df['time'] = pd.to_datetime(df['time'])
df_spi = spi(df=df, col='precip',scale=12) # here we calculate SPI
df_spi = df_spi.set_index('time') # set index for input data
```

# Now the graph of SPI


```python
fig, ax = plt.subplots(figsize=(15, 7))
ya = df_spi['spi_gamma_12']
limiar = 0 # limiar for painting the graph, where is above 0 is blue, below is red
limiar_05=-0.5
ax.fill_between(df_spi.index, ya, color='darkblue', where=ya>limiar) # fill the graph
ax.fill_between(df_spi.index, ya, color='darkred', where=ya<limiar) # fil the graph
ax.set_title('SPI 12 months', fontsize=20, fontweight='bold',fontname='Georgia') # set title
ax.set_ylabel('SPI', fontsize=16, fontweight='bold',fontname='Georgia') # set y name and font
ax.set_xlabel('Time', fontsize=16, fontweight='bold',fontname='Georgia') # set x name and font
# set limiar -0.5
ax.axhline(y=limiar_05, color='black', linestyle='--') # set limiar -0.5, when occur drought
ax.axhline(y=-limiar_05, color='black', linestyle='--') # set limiar 0.5, when occur flood
```
